#### Load The Iris Dataset

```python
from sklearn.datasets import load_iris

# Represent the iris dataset as data 
data = load_iris()

# To view the dataset remove the "#" symbol in front of data
#data

# Create variables for representing the data to be used for training
x, y = data.data, data.target
```

#### Train The Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

# Create an instance of the Decision Tree class
dtc: DecisionTreeClassifier = DecisionTreeClassifier()

# Train Decision Tree model
dtc.fit(x,y)
```

#### Basic Command Line Based Visualization

```python
from sklearn import tree

# Visualization the decission tree classifier
print(tree.export_text(dtc))  # Not very appealing to the eyes
```

#### Visualization Through Matplotlib

```python
import matplotlib.pyplot as plt

# Creating a blank canvas a.k.a figure
fig = plt.figure(figsize=(50,30))

# Create the visualization through a throw-away variable
# Note: Removing the throw-away variable initialization will display
#       the decission tree in list format alongside the visualization
_ = tree.plot_tree(dtc, feature_names=data.feature_names, class_names=data.target_names, filled=True)
```
